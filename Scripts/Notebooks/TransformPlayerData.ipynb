```python
prefix = "Season"
column_names = ["Points", "Minutes", "I", "C", "T", "Bonus", "Bonus Points", "Beginning Price", "End Price", "Goals", "Assists", "YC", "RC", "Saves", "Penalty Saves", "OG", "Penalty Misses", "CS", "GC"]
df_columns = ["Position", "Feature Year", "Target Points"] + column_names

new_df = pd.DataFrame(columns= df_columns)

new_df
```

```python
def increment_year(year: str) -> str:
    new_year = str(int(year) + 1)
    if (int(new_year) < 10):
        new_year = "0" + new_year
    return new_year

print(increment_year("6"))
```

```python
first_year = "08"
end_year = "23"
current_year = first_year
curr_dfs = []
while int(current_year) != int(end_year):
    print(current_year)
    next_year = increment_year(current_year)
    next_next_year = increment_year(next_year)
    feature_prefix = "20" + current_year + "/" + next_year + " Season "
    target_prefix = "20" + next_year + "/" + next_next_year + " Season "
    for index, row in original.iterrows():
        curr_target_points = row[target_prefix + "Points"]
        if not pd.notna(curr_target_points):
            continue
        position = row["Position"]
        curr_features = [position ,current_year, curr_target_points]
        for feature in column_names:
            feature_name = feature_prefix + feature
            curr_features.append(row[feature_name])
        curr_df = pd.DataFrame([curr_features], columns = df_columns)
        curr_dfs.append(curr_df)
    current_year = increment_year(current_year)

new_df = pd.concat(curr_dfs)
new_df = new_df.dropna()
new_df
```

```python
new_df.to_excel("FPL_Cleaned_By_Position_Until_2021.xlsx")
```
