```python
import pandas as pd
original = pd.read_excel("/Users/mlaurencechen/Library/CloudStorage/OneDrive-CornellUniversity/Internships/Projects/FPL ML 2/Data/FPL_Player_Data_SeasonUntil2024.xlsx")
original
```

```python
feature_prefix = "2022/23 Season "
column_names = ["Points", "Minutes", "I", "C", "T", "Bonus", "Bonus Points", "Beginning Price", "End Price", "Goals", "Assists", "YC", "RC", "Saves", "Penalty Saves", "OG", "Penalty Misses", "CS", "GC"]

df_columns = ["Name", "Position"] + column_names
prediction_set = pd.DataFrame(columns=df_columns)
for index, row in original.iterrows():
    name = row["Game Name"]
    position = row["Position"]
    curr_features = [name, position]
    for feature in column_names:
        feature_name = feature_prefix + feature
        curr_features.append(row[feature_name])
    prediction_set.loc[-1] = curr_features
    prediction_set.index = prediction_set.index + 1

prediction_set
```

format features and target

```python
features = pd.read_excel("/Users/mlaurencechen/Library/CloudStorage/OneDrive-CornellUniversity/Internships/Projects/FPL ML 2/Data/FPL_Cleaned_By_Position_Until_2024.xlsx")
features.drop('Unnamed: 0', axis=1, inplace= True)
defenders = features.loc[features["Position"] == "DEF"]
midfielders = features.loc[features["Position"] == "MID"]
forwards = features.loc[features["Position"] == "FWD"]
goalkeepers = features.loc[features["Position"] == "GKP"]
features = [goalkeepers, defenders, midfielders, forwards]
features
```

turn every dataframe into x_position, y_position

```python
import numpy as np
positions_x_y = []

print(defenders.columns)

for position in features:
    y = position["Target Points"]
    x = position.drop('Target Points', axis=1)
    x = x.drop('Position', axis=1)
    x = x.drop('Feature Year', axis=1)
    x = x.to_numpy()
    y = y.to_numpy()
    positions_x_y.append((x, y))

print(len(positions_x_y))
print(positions_x_y[0])
```

```python
from sklearn.model_selection import train_test_split

train_test_split_positions = []

for (x, y) in positions_x_y:
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=69)
    train_test_split_positions.append((X_train, X_test, y_train, y_test))
```

TRAINING BEGINS

```python
#random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
position_order = ["GK", "DF", "MF", "FW"]
regressors = []

for i,position in enumerate(train_test_split_positions):
    regressor = RandomForestRegressor(n_estimators = 100, random_state=30)
    regressor.fit(position[0], position[2])
    predictions = regressor.predict(position[1])
    print(position_order[i], mean_squared_error(predictions, position[3])/len(position[1]))
    regressors.append(regressor)
```

```python
position_dict = {"GKP": 0, "DEF": 1, "MID": 2, "FWD": 3}
column_names = ["Points", "Minutes", "I", "C", "T", "Bonus", "Bonus Points", "Beginning Price", "End Price", "Goals", "Assists", "YC", "RC", "Saves", "Penalty Saves", "OG", "Penalty Misses", "CS", "GC"]
predictions = []
for index, row in prediction_set.iterrows():
    if not pd.notna(row["Points"]):
        continue
    regressor = regressors[position_dict[row["Position"]]]
    data = row[column_names]
    print(data)
    data = data.to_numpy().reshape(1, -1)
    prediction = regressor.predict(data)
    predictions.append((row["Name"], prediction))

print(predictions)
```

```python
new_df = pd.DataFrame(columns= ["Name", "Predicted Points RF"])
for (player, points) in predictions:
    new_df.loc[-1] = [player, points[0]]
    new_df.index = new_df.index + 1
    
new_df.sort_values(by="Predicted Points RF", inplace = True, ascending= False)
new_df
```
